## Unemployment cleanup

```python
# read unemployment data
data1 = "../raw_data/unemployment_data1.csv"
data2 = "../raw_data/unemployment_data2.csv"
data3 = "../raw_data/unemployment_data3.csv"
data4 = "../raw_data/unemployment_data4.csv"

df1 = pd.read_csv(data1)
df2 = pd.read_csv(data2)
df3 = pd.read_csv(data3)
df4 = pd.read_csv(data4)

# add dfs to a list for merging
dfs = [df1, df2, df3, df4]

df1.head()
```

```python
# merge all dataframes
# https://stackoverflow.com/questions/23668427/pandas-three-way-joining-multiple-dataframes-on-columns
df_final = ft.reduce(lambda left,right: pd.merge(left,right,on='Unnamed: 0'), dfs)
df_final
```

```python
# delete first 9  rows of data
df_clean = df_final.iloc[9:]
df_clean
```

```python
# rename column
df_clean = df_clean.rename(columns={"Unnamed: 0": "Date"})
df_clean
```

```python
# reset index (maybe dont need it)
df_clean = df_clean.reset_index(drop=True)
df_clean.head()
```

```python
# split date to month and year
df_clean[['Month','Year']] = df_clean["Date"].str.split("-", expand=True,)
df_clean.head()
```

```python
df_clean[['Date','Month','Year']]
```

```python
# check types
df_clean.dtypes
```

```python
# Change df to string types
df_clean = df_clean.astype(str)
```

```python
# Replace NaN to 0 for columns
df_clean = df_clean.replace(np.nan, 0)
df_clean.head()
```
