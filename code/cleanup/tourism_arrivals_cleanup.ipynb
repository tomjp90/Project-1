```python
import pandas as pd
import os
```

```python
file_path = "../raw_data/travel_arrivals.csv"

# read csv and 'cut' data rows, columns and footer to desired data
data = pd.read_csv(file_path, encoding = 'ISO-8859-1', header=7, index_col = 1, skipfooter=7)
data_df = pd.DataFrame(data)

#Note, the data has been 
#(a) Country of citizenship classification and codes are from the: 
#Standard Australian Classification of Countries (SACC), 2016 (cat. no. 1269.0)
#(b) To confidentialise, estimates have been rounded to the nearest 10. As a result, sums of the components may not add exactly to totals.
data.head()
```

```python
# Transpose data
data_df = data_df.T
data_df.head()
```

```python
# Drop row 1 - 'SACC code(a)'
data_df = data_df.drop(["SACC code(a)"])
data_df
```

```python
# drop last row
data_df = data_df.iloc[:-1]
data_df
```

```python
# export data
data_df.to_csv("../output_data/tourism_arrivals_data.csv", index = False)
```
